# Combine flux tower-derived GPP files in Google Cloud

```python
!rm *.csv
!gsutil cp gs://bucketname/path-to-flux-tower-data-folder/GPP_sites_data/*.csv .
```

```python
file_list = []
for files in glob.glob("*.csv"):
  file_list.append(files)
```

```python
result_list = []
for file in file_list:
  result = pd.read_csv(file)
  result = pd.DataFrame(result, columns=['Date','GPP_PI_F_avg'])
  result['Site'] = file
  siteresult = pd.DataFrame(result)
  result_list.append(siteresult)
  result = pd.DataFrame(np.concatenate(result_list))
```

```python
result.columns = ['Date', 'GPP_measured', 'Site']
result
```

```python
result.to_csv('final_measured_GPP.csv')
!gsutil cp final_measured_GPP.csv gs://bucketname/path-to-flux-tower-data-folder/"final_measured_GPP"
```
